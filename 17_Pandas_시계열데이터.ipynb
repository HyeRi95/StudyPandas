> 시계열 데이터 분리 : strftime()

```python
now = datetime.now()
nowDate = now.strftime('%Y-%m-%d')
nowTime = now.strftime('%H:%M:%S')
nowWeek =now.strftime('%w요일')
```

```python
print(now)
print(nowDate)
print(nowTime)
print(nowWeek)
# 0:일요일, 1:월요일, 2 :화요일, 3:수요일, 4:목요일, 5:금요일, 6:토요일 
```

### 파산한 은행 수 계산 

```python
banks = pd.read_csv('C:/STUDY/StudyPandas/data/banklist.csv')
banks.info()
```

```python
banks.head()
```

```python
# 5, 6 번째 열을 datetime 오브젝트로 변환해서 읽어오기
banks = pd.read_csv('C:/STUDY/StudyPandas/data/banklist.csv',parse_dates=[5,6])
banks.info()
```

```python
banks.head()
```

- dt 접근자와 quarter(분기), year(연도) 사용

```python
# closing_quarter , closing_year 열 생성해서 값 입력 
banks['closing_quarter'] = banks['Closing Date'].dt.quarter
banks['closing_year'] = banks['Closing Date'].dt.year

banks.head()
```

```python
banks['closing_year'].value_counts().head()
```

```python
closing_year = banks.groupby(['closing_year']).size()
```

- 연도별 / 분기별 파산 현황 확인

```python
closing_year_q = banks.groupby(
    ['closing_year','closing_quarter']).size()
closing_year_q.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
closing_year.plot()
```

```python
closing_year_q.plot()
```

```python
closing_q = banks.groupby('closing_quarter').count()
closing_q
```

```python
closing_q.plot()
```

```python
plt.bar(closing_q.index,closing_q.closing_year)
```
