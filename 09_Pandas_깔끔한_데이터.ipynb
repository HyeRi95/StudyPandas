> split() 과  concat()으로 데이터 정리

```python
ebola = pd.read_csv('C:/STUDY/StudyPandas/data/country_timeseries.csv')
ebola.columns
```

```python
ebola.iloc[:5,[0,1,2,3,10,11]]
```

```python
ebola_long = pd.melt(ebola, id_vars=['Date','Day'])
ebola_long.head()
```

```python
variable_split = ebola_long.variable.str.split('_')
#ebola_long['variable']. str.split('_') 동일 
variable_split[:5]
```

```python
status_values = variable_split.str.get(0)
country_values = variable_split.str.get(1)

print(status_values[:5])
print(country_values[:5])
```

> 분리된 데이터를 ebola_long 데이터 프레임에 추가

```python
ebola_long['status']=status_values
ebola_long['country'] = country_values
ebola_long.head()
```

```python
# 두가지 방법 
variable_split = ebola_long.variable.str.split('_', expand=True)
variable_split.columns = ['status','country']
ebola_parsed = pd.concat([ebola_long, variable_split], axis=1)
ebola_parsed.head()
```

> 기상데이터의 여러열을 하나로 정리하기 

```python
weather = pd.read_csv('C:/STUDY/StudyPandas/data/weather.csv')
print(weather.shape)
print(weather.columns)
weather.iloc[:5, :11]
```

```python
weather_melt = pd.melt(weather, id_vars = ['id','year', 'month','element'],
                      var_name='day', value_name='temp')
weather_melt[100:110]
```

> pivot_table() 사용 (행 데이터를 열 데이터로)

```python
weather_tidy = weather_melt.pivot_table(index=['id','year','month','day'],
                                       columns='element', values='temp')
weather_tidy
```

```python
weather_tidy_flat = weather_tidy.reset_index()
weather_tidy_flat.head()
```

> 빌보드 차트 데이터 중복데이터 처리하기(정규화)

```python
billboard = pd.read_csv('C:/STUDY/StudyPandas/data/billboard.csv')
billboard.head()
```

```python
billboard_long = pd.melt(billboard, id_vars=['year','artist','track','time','date.entered'],
                        var_name = 'week',value_name='rating')
print(billboard_long.shape)
billboard_long.head(10)
```

```python
billboard_long[billboard_long.track == 'Loser'].head()
```

>'year' , 'artist' , 'track' , 'time'  중복데이터를 새로운 데이터 프레임으로 저장 

```python
billboard_songs = billboard_long[['year','artist','track','time']]
print(billboard_songs.shape)
```

> 중복 데이터 제거( drop_duplicates())

```python
billboard_songs = billboard_songs.drop_duplicates()
print(billboard_songs.shape)
```

> 순번(id) 추가

```python
billboard_songs['id'] = range(len(billboard_songs))
billboard_songs.head(10)
```

> 노래 정보와 주간 순위 데이터 합치기


```python
billboard_ratings = billboard_long.merge(
billboard_songs, on = ['year','artist','track','time'])
print(billboard_ratings.shape)
billboard_ratings.head()
```
