### Step 5. item_price에서 $를 없애고 float 형식으로 새롭게 저장해주세요.

```python
chipo['item_price']= chipo['item_price'].str.replace(pat='$',repl=r' ',regex=True).astype(float)
#chipo['item_price'].astype('float')
chipo
```

```python
chipo.info()
```

### Step 6. 아이템별 주문 수를 출력해주세요

```python
chipo_quantity =pd.DataFrame( chipo.groupby('item_name')['order_id'].count())
chipo_quantity
```

### Step 7. 아이템별 주문 총량을 출력해주세요.

```python
chipo_quantity = pd.DataFrame(chipo.groupby('item_name')['quantity'].sum())
chipo_quantity
```

### Step 8. 각 아이템의 가격을 구해서 unit_price 컬럼에 추가해주세요.
주문당 quantity가 1일때의 가격 (item_price / quantity)

```python
chipo['unit_price'] = chipo['item_price']/chipo['quantity']
chipo
```

### Step 9. 아이템별 개당 가격의 평균을 구하고 가장 비싼순으로 5개를 출력해주세요

```python
chipo_item_mean = pd.DataFrame(chipo.groupby('item_name')['unit_price'].mean())
chipo_item_mean.sort_values(by=['unit_price'],ascending=False).head()
```

### Step 10. 한 주문에 20달러 이상 지불한 손님을 출력해주세요.

```python
chipo_order = pd.DataFrame(chipo.groupby('order_id')['item_price'].sum())
chipo_order_20 = chipo_order[chipo_order.item_price>=20]
chipo_order_20
```
